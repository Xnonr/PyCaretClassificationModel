```python
# Import Dependencies
import pandas as pd
from pycaret.classification import * # Importing from the PyCaret classification module
```

```python
# Load Data
dataframe = pd.read_csv('heart.csv')
```

```python
dataframe.head()
```

```python
dataframe.dtypes
```

```python
# Train & Evaluate Model
categoricalFeatures = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'thal']
experiment = setup(dataframe, target = 'target', categorical_features = categoricalFeatures)
```

```python
# Trains a large number of different machine learning algorithms 
# in attempt to determine which most accurately predicts the target variable,
# ranking them as they go along as well as highlighting the best performing metrics
bestModel = compare_models()
```

```python
# Test Model
predict_model(bestModel, dataframe.drop('target', axis = 1).tail())
```

```python
# Saved Model
save_model(bestModel, model_name = 'extraTreesClassifier')
```
